**EJERCICIO:**
*Vamos a fabricar algunos datos que muestran una relación aproximadamente lineal entre la página y la cantidad de compra, la carga de la pagina es de 3 seg con una desviacion de 1 seg, con una cantidad de 100 compras*

```python
import numpy as np
import matplotlib.pyplot as plt

velPagina= np.random.normal(3.0,1.0,1000)
cantCompra= 100-(velPagina + np.random.normal(0,0.1,1000))*3

plt.scatter(velPagina,cantCompra)
```

**Como solo tenemos dos características, podemos mantenerlo simple y solo usar scipy.state.lineregress**

```python
from scipy import stats

pendiente, intercepto, r_valor, p_valor, std_err=stats.linregress(velPagina,cantCompra)
```

```python
print('No es sorprendente que nuestro valor R-cuadrado muestre un ajuste realmente bueno...', r_valor**2)
```

**Usemos la pendiente y la intersección que obtuvimos de la regresión para trazar los valores pronosticados vs observados**

```python
def prediccion(x):
    return pendiente * x + intercepto
fitline=prediccion(velPagina)
plt.scatter(velPagina,cantCompra)
plt.plot(velPagina,fitline, c='r')
plt.show()
```
