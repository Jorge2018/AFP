**¿Qué pasa si sus datos no parecen lineales en absoluto? Veamos algunos datos de velocidad pagina / compra en la página, de aspecto más realista**

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
velPagina= np.random.normal(3.0,1.0,1000)
cantCompra=np.random.normal(50.0,10.0,1000)/velPagina

plt.scatter(velPagina,cantCompra)
```

*Numpy tiene una práctica función polyfit que podemos usar, para permitirnos construir un modelo polinómico de grado n-ésimo de nuestros datos minimiza el error al cuadrado. Vamos a probarlo con un polinomio de 4º grado*

```python
x=np.array(velPagina)
y=np.array(cantCompra)

p4=np.poly1d(np.polyfit(x,y,4))
```

*Visualizaremos nuestro diagrama de dispersión original, junto con un gráfico de nuestros valores predichos usando el polinomio para tiempos de velocidad de página que oscilan entre 0 y 7 segundos.*

```python
xp=np.linspace(0,7,100)
plt.scatter(x,y)
plt.plot(xp,p4(xp),c='r')
plt.show()
print('Se ve bastante bien, midamos el error R-cuadrado')
```

```python
from sklearn.metrics import r2_score

r2=r2_score(y,p4(x))
print("El valor de R2 es...", r2)
```
