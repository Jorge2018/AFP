# A veces la precisión no cuenta toda la historia. 

*Una prueba para una enfermedad rara puede tener una precisión del 99.9% simplemente adivinando "no" todo el tiempo. Necesitamos entender los verdaderos aspectos positivos y los verdaderosnegativos, así como falsos positivos y falsos negativos. Una matriz de confusión muestra esto*

# Medicion de clasificadores

**Recall**


                *True positives*
         *____________________________*         
         *True positives + False negatives*
         
*AKA sensibilidad, tasa positiva verdadera, integridad. Porcentaje de positivos correctamente predichos. Buena elección de métrica cuando te preocupas mucho por los falsos negativos. es decir, detección de fraude*



**Precision**


                *True positives*
         *____________________________*         
         *True positives + False positives*
         
*AKA corregir positivos. Porcentaje de resultados relevantes. Buena elección de la métrica cuando te preocupas mucho por los falsos positivos. es decir, exámenes médicos, pruebas de drogas*


# Otras metricas

**Specifity**


                *True negatives*
         *____________________________*         
         *True negatives + False positives*
         
         
**F1 Score**


                *2 * True positives*
         *____________________________*         
         *2 * True positives + False positives + False negatives*
         
         
              *2  * Precision * Recall *
         *____________________________*         
              *2  * Precision + Recall *
         
*Armónico medio de precisión y sensibilidad. Cuando te preocupas por la precisión y el recall*


**RMSE**
*Error cuadrático medio de la raíz, exactamente como suena. Medición de precisión. Solo se preocupa por las respuestas correctas e incorrectas*


# Curva ROC

*Curva característica de funcionamiento del receptor. Gráfico de tasa de verdaderos positivos (recuperación) frente a tasa de falsos positivos en varios umbrales. Los puntos por encima de la diagonal representan una buena clasificación (mejor que al azar). La curva ideal sería solo un punto en la esquina superior izquierda. Cuanto más se "incline" hacia la parte superior izquierda, mejor*

# AUC

*el área bajo la curva ROC es.. Espéralo. Área bajo la curva (AUC). Igual a la probabilidad de que un clasificador clasifique una instancia positiva elegida al azar más alta que una instancia negativa elegida al azar. ROC AUC de 0.5 es un clasificador inútil, 1.0 es perfecto. Métrica comúnmente utilizada para comparar clasificadores*
