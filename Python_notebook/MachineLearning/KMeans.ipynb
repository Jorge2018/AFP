Iterar hasta que los puntos dejen de cambiar la asignación a centroides. Si desea predecir el clúster para nuevos puntos, simplemente encuentre el centroide al que están cercados*

**k-significa agrupar gotchas**

*Elección de K. Intente aumentar los valores de K hasta que deje de obtener grandes reducciones en el error al cuadrado (distancias desde cada punto hasta sus centroides). Evitar los mínimos locales. La elección aleatoria de los centroides iniciales puede producir resultados diferentes. Ejecútelo varias veces solo para asegurarse de que su resultado inicial sea el esperado. Etiquetado de los clústeres. K-mean no intenta asignar un ningun significado a los clústeres que encuentre. Depende de usted profundizar en los datos y tratar de determinar que se trata*

**Ejemplo**

*Hagamos algunos datos falsos que incluyan personas agrupadas por ingresos y edad, al azar.*

```python
import numpy as np

def creaData(N,k):
    np.random.seed(10)
    puntosCluster= float(N)/k
    x=[]
    for i in range (k):
        ingresos=np.random.uniform(20000.0,200000.0)
        edad=np.random.uniform(20.0,70.0)
        for j in range (int(puntosCluster)):
            x.append([np.random.normal(ingresos,10000.0), np.random.normal(edad,2.0)])
    x=np.array(x)
    return x
```

**Usaremos KMeans para redescubrir estos clústeres en el aprendizaje no supervisado**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

data=creaData(100,5) # crea arreglo de 100 personas al azar en 5 grupos

modelo= KMeans(n_clusters=5)

modelo=modelo.fit(scale(data))

print (modelo.labels_)

plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1], c=modelo.labels_.astype(np.float))
plt.show()
```

**Ejercicio con diversos centroides**

```python
import warnings
warnings.filterwarnings("ignore")

lstcentroides=[3,5,7,11,15]

for i in lstcentroides:
    data=creaData(100,i) # crea arreglo de 100 personas al azar en ? grupos
    modelo= KMeans(n_clusters=i)
    modelo=modelo.fit(scale(data))
    plt.figure(figsize=(6,4))
    plt.scatter(data[:,0],data[:,1], c=modelo.labels_.astype(np.float))
    print(f'Grafo que tiene un total de {i} agrupaciones..')
    plt.show()
```
