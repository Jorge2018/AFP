```python
import pandas as pd
import numpy as np
from ast import literal_eval
import re

rating=pd.read_csv(r'./ratings_small.csv')
movie=pd.read_csv(r'./movies_metadata.csv')

regex = re.compile(r'[^0-9]') # Eliminamos todo lo que no sean números
movie['id']=movie['id'].replace(regex, '').astype(int)
ratingok=rating.merge(movie,left_on='movieId',right_on='id')
ratingok=ratingok[['movieId','belongs_to_collection']]
ratingok=ratingok.dropna()

ratingok.belongs_to_collection = ratingok.belongs_to_collection.fillna('{}')
ratingok.belongs_to_collection = ratingok.belongs_to_collection.apply(literal_eval)

ratingok = ratingok.join(pd.json_normalize(ratingok.pop('belongs_to_collection')))
ratingok=ratingok.dropna()
ratingok = ratingok[~ratingok.astype(str).duplicated()]
ratingok = ratingok[['movieId','id','name']]

ratingok=pd.merge(ratingok, rating, how = 'left', on = None, left_on = ['movieId'],   # match entre df para crear otro df
                 right_on = ['movieId'], left_index = False, right_index = False)
ratingok = ratingok[~ratingok.astype(str).duplicated()]
ratingok
```

```python
userRating=ratingok.pivot_table(index='userId',columns=['name'],values='rating')
userRating.head()
```

```python
corrMatriz=userRating.corr()
corrMatriz.head()
```

**Sin embargo, queremos evitar los resultados de spurios que ocurrieron de solo un puñado o usuarios que calificaron el mismo par de películas. Con el fin de restringir nuestros resultados a películas que mucha gente calificó como más altas, y también darnos resultados más populares que son más fácilmente reconocibles, usaremos el argumento min_periods para arrojar nuestros resultados donde menos de 100 usuarios calificaron un par de películas determinado.**

```python
corrMatriz=userRating.corr(method='pearson', min_periods=100)
corrMatriz.head()
```

**Ahora vamos a producir algunas recomendaciones de películas para el ID de usuario 1, A este tipo realmente le gustan 'American Pie Collection' y 'The Space Odyssey Series', pero odiaba 'Rocky Collection'. Extraeré sus calificaciones del dataframe userRatings y usaré dropna() para deshacerme de los datos faltantes (dejando mi solo con una serie de las películas que realmente califiqué)**

```python
miRank=userRating.loc[1].dropna()
miRank
```

**Ahora revisemos cada película que califiqué una a la vez, y construyamos una lista de posibles recomendaciones basadas en las películas similares a las que califiqué. Entonces, para cada película que califiqué, recuperaré la lista de películas similares de nuestra matriz de correlación. Luego escalaré esos puntajes de correlación por lo bien que califiqué la película a la que son similares, por lo que las películas similares a las que me gustaron cuentan más que las películas similares a las que odié.**

```python
simCandidato=pd.Series()
for i in range(0,len(miRank.index)):
    print('Agregando similar para...'+miRank.index[i]+'....')
    similar=corrMatriz[miRank.index[i]].dropna()
    similar=similar.map(lambda x:x * miRank[i])
    simCandidato=simCandidato.append(similar)
    
    
print('Ordenando....')
simCandidato.sort_values(inplace=True, ascending=False)
simCandidato.head(10)
```

**¡Esto está empezando a parecer algo útil! Tenga en cuenta que algunas de las mismas películas aparecieron más de una vez, porque eran similares a más de una película que califiqué. Usaremos group by() para agregar más de las puntuaciones de las películas que aparecen más de una vez, por lo que contarán más**

```python
simCandidato=simCandidato.groupby(simCandidato.index).sum()
simCandidato.sort_values(inplace=True,ascending=False)
simCandidato.head(10)
```

**lo último que tenemos que hacer es filtrar las películas que ya he calificado, ya que recomendar una película que ya he visto no es útil**

```python
filtro=simCandidato.drop(miRank.index)
simCandidato.head(10)
```
