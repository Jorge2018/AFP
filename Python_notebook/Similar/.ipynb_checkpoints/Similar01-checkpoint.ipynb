# Buscando peliculas similares

*Comenzaremos cargando el conjunto de datos de MovieLens. Usando pandas, podemos cargar muy rápidamente las filas de los archivos u.data y you.item que nos interesan, y fusionarlos para que podamos trabajar con nombres de películas en lugar de ID. (En un trabajo de producción real, te quedarías con las identificaciones y te preocuparías por los nombres en la capa de visualización para hacer las cosas más eficientes. Pero esto nos permite entender lo que está pasando mejor por ahora)*

```python
import pandas as pd
import numpy as np
from ast import literal_eval
import re

rating=pd.read_csv(r'./ratings_small.csv')
movie=pd.read_csv(r'./movies_metadata.csv')

regex = re.compile(r'[^0-9]') # Eliminamos todo lo que no sean números
movie['id']=movie['id'].replace(regex, '').astype(int)
ratingok=rating.merge(movie,left_on='movieId',right_on='id')
ratingok=ratingok[['movieId','belongs_to_collection']]
ratingok=ratingok.dropna()

ratingok.belongs_to_collection = ratingok.belongs_to_collection.fillna('{}')
ratingok.belongs_to_collection = ratingok.belongs_to_collection.apply(literal_eval)

ratingok = ratingok.join(pd.json_normalize(ratingok.pop('belongs_to_collection')))
ratingok=ratingok.dropna()
ratingok = ratingok[~ratingok.astype(str).duplicated()]
ratingok = ratingok[['movieId','id','name']]

ratingok=pd.merge(ratingok, rating, how = 'left', on = None, left_on = ['movieId'],   # match entre df para crear otro df
                 right_on = ['movieId'], left_index = False, right_index = False)
ratingok = ratingok[~ratingok.astype(str).duplicated()]
ratingok
```

**Ahora la increíble función pivot_table en un marco de datos construirá una matriz de clasificación de usuario / película. Observe cómo NAN indicó datos faltantes: películas que el usuario específico no calificó**

```python
movieRating=ratingok.pivot_table(index='userId',columns=['name'],values='rating')
movieRating.head()
```

**Vamos a extraer series de usuarios que calificaron American Pie Collection**

```python
ameriP=movieRating['American Pie Collection']
ameriP.head()
#movieRating.info()
```

**¡La función Pandas corrwith hace que sea realmente fácil calcular el vector de correlación por pares 'American Pie Collection' de la calificación del usuario con cualquier otra película! Después de eso, eliminaremos cualquier resultado que no tenga datos y construiremos un nuevo marco de datos de películas y su puntaje de correlación (similitud) con 'American Pie Collection'**

```python
peliSimilar=movieRating.corrwith(ameriP)
peliSimilar=peliSimilar.dropna()
df=pd.DataFrame(peliSimilar)
df.head(15)
```

**Ordenemos el resultado por puntuación de similitud, y deberíamos tener las películas más similares a 'American Pie Collection' Excepto... Nosotros no. ¡Estos resultados tienen sentido! Es por eso que es importante conocer sus datos: claramente nos perdimos algo importante**

```python
peliSimilar.sort_values(ascending=False)
```

**Nuestros resultados probablemente se están estropeando con películas que solo han sido vistas por un puñado de personas a las que también les gustó 'American Pie Collection'. Así que tenemos que deshacernos de las películas que sólo fueron vistas por unas pocas personas que están produciendo resultados espurios. Construyamos un nuevo marco de datos que cuente cuántas calificaciones existen para cada película y también la calificación promedio mientras estamos en ello, eso también podría ser útil más adelante.**

```python
peliStats=ratingok.groupby('name').agg({'rating':[np.size,np.mean]})
peliStats.head()
```

**Deje que Grid se deshaga de cualquier película calificada por menos de 100 personas y verifique las mejor calificadas que queden**

```python
peliPopular=peliStats['rating']['size']>=100
peliStats[peliPopular].sort_values([('rating','mean')],ascending=False)[:15]
```

**100 aún puede ser demasiado bajo, pero estos resultados se ven bastante bien en cuanto a "películas bien calificadas de las que la gente ha oído hablar". Unamos estos datos con nuestro conjunto original de películas similares a 'American Pie Collection'**

```python
df=peliStats[peliPopular].join(pd.DataFrame(peliSimilar, columns=['similarity']))
df.head()
```

**y, ordenar estos nuevos resultados por puntaje similitar. Ese es un modo como este**

```python
df.sort_values(['similarity'],ascending=False)[:15]
```
